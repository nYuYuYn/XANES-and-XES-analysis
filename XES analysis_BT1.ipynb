```python
path = r'C:\Users\oppcexpv\Yuan-Chi code\XES_fit\csv'
path = path.replace('\\', '/')
i = '2b_XES_fit.csv'

test_file = pd.read_csv(path+'/'+i)

for j in test_file.columns[4:]:
    test_file[j] - test_file['bg']
A1 = integrate.simpson(test_file['p1'])
A2 = integrate.simpson(test_file['p2'])
A3 = integrate.simpson(test_file['p3'])
A4 = integrate.simpson(test_file['p4'])
ratio = (A1+A2)/(A3+A4)
print(A1, A2, A3, A4, ratio)
```

```python
### Intergrate the peak by brut force


path = r'C:\Users\oppcexpv\Yuan-Chi code\XES_fit\csv'
path = path.replace('\\', '/')

file_name = []
dfs = []
for i in os.listdir(path):
    file_name.append(i)
    dfs.append(pd.read_csv(path+'/'+i))

Area_ratio = []
for i in range(len(file_name[:-2])):
    x = dfs[i]['x']
    Area = []
    for j in dfs[i].columns[4:]:
        dfs[i][j] - dfs[i]['bg']
        Area.append(integrate.simpson(dfs[i][j], x))
    ratio = (Area[0]+Area[1])/(Area[2]+Area[3])
    Area_ratio.append(ratio)
```

```python
for i in range(len(file_name[:-2])):
    print(file_name[i], Area_ratio[i])
```

```python
PSTFSI_con = [0.061,0.029,0.013,0.039,0.032,0.015]
PSTFSI_ratio = Area_ratio[:6]
EG_TFSI_con = [0.062,0.031,0.016,0.063,0.032,0.015]
EG_TFSI_ratio = Area_ratio[6:]
```

```python
fig = plt.figure()

plt.plot( PSTFSI_con[:3],PSTFSI_ratio[:3], label='Li2S8')
plt.plot( PSTFSI_con[3:],PSTFSI_ratio[3:], label='Li2S4')

plt.xlabel('Concentration (R)')
plt.ylabel('A(Li2Sx) / A(TFSI)')
plt.title('PSTFSI')
plt.grid()
plt.legend()

plt.savefig('PSTFSI_XES_peak ratio')
```
